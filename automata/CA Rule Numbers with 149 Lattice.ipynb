Visualizing the State Space Diagrams of all Rule Numbers

```python
import cellpylib as cpl
import numpy as np
import matplotlib.pyplot as plt
```

```python

rule_numbers = {
    "GKL": 493275728351097185153566218236460895,
    "GP": 6647864381429528876515014584928523775,
    "GEP_1": 22598599938439660964756895944266438143,
    "GEP_2": 441354661229560436815860362649141111,
    "DAVIS": 244106326293841510796857170733301535,
    "DAS": 9304758155185428497243733980536180735,
    "MM401": 339841014795010104073313096675879420072,
    "TEMP" : 1
}



bin = '11111111001010101111111110000000111111110000000011111111100000001111111110001010111111111000100001000100000000000101010000001000'
print(int(bin, 2))
rule_numbers["TEMP"] = int(bin, 2)
```

```python
cellular_automaton = cpl.init_random(149)
noise_prob = 0.0

automaton = {}
item = "GP"
rule_number = rule_numbers[item]

def function_noise(n, c, t):
    n_to_use = n.copy()
    for i,x in enumerate(n_to_use):
        if np.random.random() < noise_prob:
            n_to_use[i] = 1-x
    return cpl.binary_rule(n_to_use, rule_number)


automaton[item] = cpl.evolve(
    cellular_automaton, timesteps=200, apply_rule=function_noise, r=3
)
print(f"{item} Rule")
plt.figure(figsize=(10, 5))
cpl.plot(automaton[item])

```
